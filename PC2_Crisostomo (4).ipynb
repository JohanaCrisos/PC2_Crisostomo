```python
import pandas as pd
link='https://github.com/JohanaCrisos/PC2_Crisostomo/raw/refs/heads/main/HDI.xlsx'
tabla=pd.read_excel(link, header=4)
tabla
```

```python
tabla.columns
```

```python
#Eliminar filas y columnas innecesarias
toDrop=['Unnamed: 0','Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13','GNI per capita rank minus HDI rank', 'HDI rank']
tabla.drop(columns=toDrop, inplace=True)
tabla.drop(index=[0,1], inplace=True)
```

```python
#quitar los espacios iniciales y finales de las columnas y sustituir
#los espacios entre lineas con _
tabla.columns=tabla.columns.str.strip().str.replace(r'\s+', '_', regex=True)
tabla.columns
```

```python
#Renombrar
newname={'Unnamed:_1':'Country','Human_Development_Index_(HDI)':'Human_Development_Index','Gross_national_income_(GNI)_per_capita':'Gross_national_income_per_capita'}
tabla.rename(columns=newname,inplace=True)
```

```python
tabla
```

```python
#Eliminar filas con datos faltantes
tabla=tabla.dropna(axis='rows',how='any', ignore_index=True)
tabla
```

```python
#Eliminar las final que no tengan nada que ver con Country
tabla=tabla.drop(index=range(195,210))
tabla
```
